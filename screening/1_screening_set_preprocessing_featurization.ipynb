```python
from rdkit.ML.Descriptors import MoleculeDescriptors
def compute_2Drdkit(mol):
    calc = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
    ds = calc.CalcDescriptors(mol)
    return list(ds)


#Morgan2 FP
def get_morgan2(mol):
    return list(AllChem.GetMorganFingerprintAsBitVect(mol,2,nBits=2048))
```

```python
for i in range(7,10):
    df = pd.read_csv(f'/home/sliu/cyp_related_coding/RF_CYP19/processing_vitasmdb/vitasm_preprocessed{i}_filtered.csv')
    print(f'read csv{i}')
    #rdkit 2D descriptors
    PandasTools.AddMoleculeColumnToFrame(df,'preprocessedSmiles','Molecule')
    descriptors_list = [x[0] for x in Descriptors._descList]
    df['descriptor'] = df.Molecule.apply(compute_2Drdkit)
    ##one descriptor per column
    calc = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
    desc_cols = list(calc.GetDescriptorNames())
    df[desc_cols] = df.descriptor.to_list()
    df[desc_cols].dropna(inplace=True)
    df.reset_index(inplace=True)
    #Morgan2 FP
    df['fp'] = df.Molecule.apply(get_morgan2) #doesn't work if save the fp in one column
    morgan2_cols = ['morgan2_b'+str(i) for i in list(range(2048))]
    df[morgan2_cols] = df.fp.to_list()
    df.to_csv(f'/home/sliu/cyp_related_coding/RF_CYP19/processing_vitasmdb/vitasm_preprocessed_filtered_featrized{i}.csv', index=False)
```
